```python
# Categorize columns into categorical and numerical based on dtype

# Categorical columns (dtype == 'object')
cat_cols = [col for col in df.columns if df[col].dtype == 'object']

# Numerical columns (dtype != 'object')
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# Load the dataset
df_training = pd.read_csv('/kaggle/input/disease-prediction/Training.csv')
df_testing = pd.read_csv('/kaggle/input/disease-prediction/Testing.csv')

# List of features (symptoms columns)
feature_columns = [
    'itching', 'skin_rash', 'nodal_skin_eruptions', 'continuous_sneezing', 'shivering',
    'chills', 'joint_pain', 'stomach_pain', 'acidity', 'ulcers_on_tongue'
]

# Train and test split (using the appropriate feature columns)
X_train = df_training[feature_columns]  # Symptoms/features
y_train = df_training["prognosis"]  # Target (disease label)
X_test = df_testing[feature_columns]  # Symptoms/features
y_test = df_testing["prognosis"]  # Target (disease label)

# Encode labels (if needed) for target variable
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# -------------------------------------------------------------------------------------------------------------------
# Decision Tree Model
clf_dt = DecisionTreeClassifier()
clf_dt.fit(X_train, y_train)

# Predictions
y_pred_dt = clf_dt.predict(X_test)

# Metrics for Decision Tree
print("Decision Tree Metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_dt):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_dt, average='weighted', zero_division=1):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_dt, average='weighted'):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_dt, average='weighted'):.4f}")

# ROC AUC (using one-vs-rest strategy for multi-class classification)
y_prob_dt = clf_dt.predict_proba(X_test)  # Get the probability for each class
print(f"ROC AUC: {roc_auc_score(y_test, y_prob_dt, multi_class='ovr', average='weighted'):.4f}")

# -------------------------------------------------------------------------------------------------------------------
# Random Forest Model
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)

# Predictions
y_pred_rf = clf_rf.predict(X_test)

# Metrics for Random Forest
print("\nRandom Forest Metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf, average='weighted', zero_division=1):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_rf, average='weighted'):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_rf, average='weighted'):.4f}")

# ROC AUC (using one-vs-rest strategy for multi-class classification)
y_prob_rf = clf_rf.predict_proba(X_test)  # Get the probability for each class
print(f"ROC AUC: {roc_auc_score(y_test, y_prob_rf, multi_class='ovr', average='weighted'):.4f}")

# -------------------------------------------------------------------------------------------------------------------
# Naive Bayes Model
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predictions
y_pred_gnb = gnb.predict(X_test)

# Metrics for Naive Bayes
print("\nNaive Bayes Metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_gnb):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_gnb, average='weighted', zero_division=1):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_gnb, average='weighted'):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_gnb, average='weighted'):.4f}")

# ROC AUC (using one-vs-rest strategy for multi-class classification)
y_prob_gnb = gnb.predict_proba(X_test)  # Get the probability for each class
print(f"ROC AUC: {roc_auc_score(y_test, y_prob_gnb, multi_class='ovr', average='weighted'):.4f}")
```
